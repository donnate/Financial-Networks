<h1>  <center> Hierarchical Clustering </center> </h1>

<span style='color:blue'> __*Quick recap on the project's objectives*__</span> The idea here is to try to build a graph between stocks, capturing their inter-dependencies and similarities. We work on the log-returns: denoting as $P_{st}$ the price at time $t$ of stock $s$: $R_{st}=\log(\frac{ P_{s,t+1}}{P_{st}})$.

(Traditional assumptions on that quantity are that for every stock, these returns are independent from day to day and follow a normal distribution).

<span style='color:purple'>  __The model__:
$$\forall s \in \mathcal{S}, \forall t\in T, \quad R_{st}=\beta_sR_t^{(M)} +\epsilon_{st}$$ </span> 
where $R^{(M)}$ denotes the return of the market (that is, an overall trend. $\epsilon_{st}$ is the stock residual (that is the projection of the time series of each stock in the space orthogonal to the market trend). The idea is that stocks might be correlated-- that is, beyond the market trend that they all more or less closely follow, and we wish to evaluate these correlations, and encapsulate the information about these interdependencies in graph.



__First things first: Why is this an interesting problem?__
+ *Why do we need a graph:* in general, the idea is to divide the market into clusters of stocks which exhibit correlated behaviors. This clustering is then used either in risk-evaluation or for optimal $\alpha$ prediction.: stocks clustered together are highly correlated, and in order to diversify her portfolio and to hedge her risk, the investor will typically try to invest in different "pools"/clusters of stocks. Graphs can be considered as an extension to basic clusters: they encapsulate finer grain information about the distance between graphs than clusters (which are basically disconnected complete graphs).  __The main goal of clustering/ summarizing the correlation matrix with a graph is to filter the correlation matrix, and to assess which stocks are truly correlated amidst all that noise__
+ *What should make our proposed approach better?* We wish to make the graph framewwork a more robust and with more snese than most of what has been previously achieved: we do not want to fit an unstable structure like the MST that are going to change from time period to time period. 


The idea here is to try to cluster stocks, using a bottom-up "hierarchical clustering" based method:
+ at every level of the tree, we should be able to recover the graph corresponding to any  level of thresholding of the correlation matrix: we still want to find a way to easily go back to the clustering that is typically being realised (IMR style approach: the tree will give us a multu-scale view of the network.)
+ The binning of the correlations might induce more robustness  to noise as well


*__ Random Notes gathered up from reading__*:
+ including information about distributions in our metric also makes sense, especially from an investment perspective (among clusters, we want to cluster stocks by volatility for instance.) Hence the extension of the project that was concerned with adding information to the distance might be a good direction.
+ Pearson correlations are easier (i.e, faster to compute) than the others, but they are not a good metric!!! Especially if we are comparing time-series that do not follow normal distributions and such. The best is allegedly Kendall's tau (to be looked into). However, Spearman's correlation is good as well: it has an interpretation in terms of copula transforms: basically, computing Spearman's correlation is euqivalent to computing Pearson's correlation on the transformed copulas. It experimentally yields better results. So perhaps we should go down this way as well.

<h2> <span style="color:maroon"> II. Proposed Approach </span> </h2>

We want to filter the correlation using a tree based structure, binning correlations at every level.

<span style='color:blue'>__Detailed Proposed Approach__</span>:
+ Start with N nodes, and a step\_size
+ for rho in [1:: -step_size :: 0]:
    + find  all pairs of nodes $i$ and $j$ such that $\rho_{ij} \in [rho, rho+step\_size]$ and associate them to a cluster. If there exists a cluster at that level of correlation that already contains $i$ or $j$, add the new value to that cluster.
    + once all clusters at level $rho$ have been created, create edges between the nodes of the clusters.
    + For each of the clusters (which contain one or more values), define a "super node"-- that is, a parent of all the nodes within that clusters, and connect it to its "children"
    + Define a weight for the $k^{th}$ super node at level $rho$ associated to cluster $C_{rho,k}$ as:
    
    $$ w_{rho,k}=\sum_{stock \in C_{rho,k}} w_{stock}$$
    + Replace the stocks  by the weighted time series of their super node defined as:
    $$R_{st}^{C_{rho,k}}=\frac{1}{\sum_{stock \in C_{rho,k}} w_{stock}}\sum_{stock \in C_{rho,k}} w_{stock} R_{stock,t}$$
    + Iterate the process until with the "smaller" correlation matrix

<span style='color:blue'>__Useless Computations to try to make sense of it all__</span>:
At a higher level (smaller correlation) in the tree:
    $$ \text{Corr}(w_a X_a+ w_b X_b, X_c)=\frac{ w_a\sigma_a}{\sqrt{w_a^2 \sigma_a^2+w_b^2 \sigma_b^2+2 w_a w_b \sigma_a \sigma_b \rho_{ab}}} \text{Corr}(X_a, X_c)+\frac{ w_b\sigma_b}{\sqrt{w_a^2 \sigma_a^2+w_b^2 \sigma_b^2 +2 w_a w_b \sigma_a \sigma_b \rho_{ab}}} \text{Corr}(X_b, X_c)$$
    $$ \text{Corr}(w_a X_a+ w_b X_b, X_c)=\frac{ 1}{\sqrt{1 +w _r^2 +2 w_r  \rho_{ab}}} \text{Corr}(X_a, X_c)+\frac{w_r}{\sqrt{1+w_r^2  +2 w_r \rho_{ab}}} \text{Corr}(X_b, X_c)$$
    with $w_r=\frac{w_b \sigma_b}{w_a \sigma_a}$
    $$ \implies \text{Corr}(w_a X_a+ w_b X_b, X_c)=\frac{ 1}{ \sqrt{(1 +w _r)^2 -2 w_r (1- \rho_{ab})}} \text{Corr}(X_a, X_c)+\frac{w_r}{\sqrt{(1+w_r)^2  -2 w_r (1-\rho_{ab})}} \text{Corr}(X_b, X_c)=\frac{1}{\sqrt{1  +2 \frac{w_r}{(1 +w _r)^2}  (1-\rho_{ab})}}  \Big[\frac{1}{1+w_r} \text{Corr}(X_a, X_c)+\frac{w_r}{w_r+1} \text{Corr}(X_b, X_c) \Big]$$
    with $w_r=\frac{w_b \sigma_b}{w_a \sigma_a}$

Suppose that:
$\text{Corr}(X_a,X_c) >\text{Corr}(X_b,X_c)$, then:
$$\text{Corr}(w_a X_a+ w_b X_b, X_c) \leq \frac{1}{\sqrt{1  +2 \frac{w_r}{(1 +w _r)^2}  (1-\rho_{ab})}}  \rho_{ac} \leq \rho_{ac} \leq \rho_{ab}$$
where the last ineuality follows by design of the algorithm.
Hence the correlation between the super node and each remaining stock induces correlations that are necessarily monotonous, and an ordering of the distances that is consistent with the one given by the original matrix.

Hence the super nodes are weighted barycenters of the nodes that they contain. Since the time series of a super node ofanother barycenter yieldS:
    

$$ R_{st}=\frac{1}{\sum_{C_k \in C^{++}}\sum_{s \in C_k} w_s } \sum_{C_k \in C^{++}} (\sum_{s \in C_k} w_s ) \times \sum_{s \in C_k}  [\frac{1}{\sum_{s \in C_k} w_s} \sum_{s in c_k }w_s R_{st}]$$
Each super node is in fact the barycenter of all nodes conatined below with respect to their original weights. Hence, each super node contains information about a cluster that wuld have been achieved if we had thresholded correlations below rho.


<h2> <span style="color:maroon"> III. The road so far</span> </h2>

We have to investigate how well the tree that is obtained by this method behaves. To introduce more stability, we propose to:
+ use spearman's correlation instead of Pearson
+ bootstrap the correlation, compute the average over all bootstrap samples, and keep the ones that are truly significant. 

We do this for periods of one year.

<span style='color:blue'>__Proposed evaluation of our results__</span>:
+ stability of the graph over time
+  good partial matching of our clusters with the Sectors/ Industries clustering


### a. Assessing the fit of the  Bootstrap distribution

<span style='color:blue'>__*Objective*__</span>: Here, we want to __induce more stability in the graph by working on a "bagged" bootstrapped distribution__: we draw several bootstrap correlation matrices, and average them. The idea is  that the average bagged correlation matrice that we thus obtain should converge to the "true" correlation matrix. 

<span style='color:blue'>__*Problem*__</span>: For the previous to hold, we have to ensure that __the bootstrap samples truly do come from the same distribution as the observed correlation matrix__. In particular, we usually suppose that __the returns are independent from day to day, but this assumption might not be accurate__. The bootstrap samples might thus exhbit a slightly different behavior if we do not take this into account. Hence, we could prefer block-bootstrapping. This raises the question of the __optimal block size__.

<span style='color:blue'>__*Proposed Evaluation procedure*__</span>: 
We want to assess if the observed correlation matrix is a plausible observation amidst the bootstrap samples-- that is, if the observe correlation matrix and the bootstrap samples are generated by the same distribution. The problem is that there might be some correlations between coefficients: if one correlation is unsually high with respect to the other bootstrap samples, then it might be the case that is it also registered as significantly different in other pairs as well.  To correct for that effect, we could try to apply a Benjamini and Hochberg procedure to try to determine the true number of significantly different correlations (as measured by their empirical p-value), as detailed below:

For each suggested block size $b$:
+ (1) Compute the block-bootstrap distibution 
+ (2) Assess the number of observed correlations that lie outside of the 95% confidence interval. 
+ (3) Reject those that are below the BH threshold under correlation. That gives an estimate of the number of outlier values

```python
y=1995
B=200
from tree_attempt import *
S=[1,2,3,4,5,6,7]
import scipy as sc
path='/Users/cdonnat/Dropbox/'
print 'starting year ',y
stock_data=pickle.load(open(path+"data/stock_data"+str(y)+'.pkl','rb'))
stock_residuals=pickle.load(open(path+'data/stock_res'+str(y)+'.pkl','rb'))
Volume_data=pickle.load(open(path+'data/volume_data'+str(y)+'.pkl','rb'))
weights=compute_weights(stock_data,Volume_data)
cor_res, pval = sc.stats.spearmanr(stock_residuals)
cor_res = pd.DataFrame(cor_res, index=stock_residuals.columns, columns=stock_residuals.columns)
sig={}
for s in S:
    print('s is '+str(s))
    dict_cor=pd.DataFrame(cor_res.values.reshape([-1,1]))
    dict_cor.columns=[0]
    index_block=np.arange(stock_residuals.shape[0])//s
    for b in range(B):
            print(b)
            bdays=np.random.choice(range(stock_residuals.shape[0]//s),stock_residuals.shape[0]//s+(s-1),replace=True)
            bootstrap_days=[]
            for block in bdays:
                bootstrap_days+=list(stock_residuals.index[index_block==block])
            bootstrap_days=bootstrap_days[:stock_residuals.shape[0]]    
            bootstrap_residuals=stock_residuals.loc[bootstrap_days]
            boot_cor=bootstrap_residuals.corr()
            boot_cor, pval = sc.stats.spearmanr(bootstrap_residuals)
            boot_cor = pd.DataFrame(boot_cor, index=bootstrap_residuals.columns, columns=bootstrap_residuals.columns)
            dict_cor[b+1]= boot_cor.values.reshape([-1,1])
    pickle.dump(dict_cor,open('/Users/cdonnat/Dropbox/data/dict_cor1995_'+str(s)+'.pkl','wb'))
    summary_q975=dict_cor.iloc[:,1:].quantile(0.975,1)
    summary_q025=dict_cor.iloc[:,1:].quantile(0.0275,1)
    summary_var=dict_cor.iloc[:,1:].std(1)
    print(summary_var.shape)
    cap_down=pd.DataFrame(dict_cor[0]-summary_q025)
    cap_up=pd.DataFrame(summary_q975-dict_cor[0])
    significantly_diff=(cap_up<0)|(cap_down<0)
    ### Rank the observations to get the p-values
    pval=1.0/(B+2)*(dict_cor.rank(1)).iloc[:,0]
    index_pval=pval.index
    N=len(pval)
    pval.index=range(N)
    alpha=0.05
    pvalBH=pval.values-np.array([alpha*k*1.0/N for k in range(N)])
    ## nb significant:
    test=np.where(pvalBH<0)
    if len(test[0])==0:
        sig[s]=0
    else:
        sig[s]=np.max(test[0])
    print 'sig is ',sig
    pval.index=index_pval
    pickle.dump(significantly_diff,open('/Users/cdonnat/Dropbox/data/sig_diff1995_'+str(s)+'.pkl','wb'))
    pickle.dump(pval,open('/Users/cdonnat/Dropbox/data/pval1995_'+str(s)+'.pkl','wb'))
```
